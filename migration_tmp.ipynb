```python
from pathlib import Path

root = Path('/workspaces/PROMPTBOOK')
excluded_dirs = {'.git', '.venv', 'node_modules', '__pycache__'}
marker = '## FILE-SPECIFIC-EXPANSION'

cleaned = []
for path in sorted(root.rglob('*.md')):
    if any(part in excluded_dirs for part in path.parts):
        continue

    text = path.read_text(encoding='utf-8')
    pos = text.find(marker)
    if pos == -1:
        continue

    path.write_text(text[:pos].rstrip() + '\n', encoding='utf-8')
    cleaned.append(str(path.relative_to(root)))

print('CLEANED FILES:', len(cleaned))
for rel in cleaned:
    print(rel)
```

```python
from pathlib import Path

root = Path('/workspaces/PROMPTBOOK')
excluded_dirs = {'.git', '.venv', 'node_modules', '__pycache__'}

rows = []
for path in sorted(root.rglob('*.md')):
    if any(part in excluded_dirs for part in path.parts):
        continue
    with path.open('r', encoding='utf-8') as f:
        lines = sum(1 for _ in f)
    rows.append((str(path.relative_to(root)), lines))

print('FILES:', len(rows))
for rel, lines in rows:
    print(f'{lines:5d}  {rel}')
```
